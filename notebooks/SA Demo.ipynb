```python
import nltk

nltk.download('punkt')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import tokenize

def print_polarity(ss):
    print('> ', end='')
    for k in sorted(ss):
        print('{0}={1} '.format(k, ss[k]), end='')
    print()

tweets = [
         """
         We have the greatest economy anywhere in the world. 
         We have the greatest military anywhere in the world. Not bad!
         """,
         """
         Iran was violating the 150 Billion Dollar (plus 1.8 
         Billion Dollar in CASH) Nuclear Deal with the United 
         States, and others who paid NOTHING, long before I became 
         President - and they have now breached their stockpile 
         limit. Not good!
         """,
         "Iran made a very big mistake!",
         """
         Great news for the Republican Party as one of the dumbest & 
         most disloyal men in Congress is “quitting” the Party. No 
         Collusion, No Obstruction! Knew he couldn’t get the nomination 
         to run again in the Great State of Michigan. Already being 
         challenged for his seat. A total loser!
         """]
sid = SentimentIntensityAnalyzer()

for tweet in tweets:
     sentences = tokenize.sent_tokenize(tweet)
    
     for sentence in sentences:
         print(sentence)
         ss = sid.polarity_scores(sentence)
         print_polarity(ss)
    
     print()
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
for tweet in tweets:
    annotated = nlp(tweet)
    
    print(tweet)
    print()
    
    nsubj = set()
    
    for t in annotated:
        if not t.text.strip():
            continue
        
        if t.dep_ is 'nsubj':
            nsubj.add(t.text)
            
        print("{} = {} ({})".format(t, t.tag_, spacy.explain(t.tag_)))
        
    print()
    print("Nominal subjects: {}\n".format(nsubj))
```
